```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('/content/MGN_ADM_MPIO_GRAFICO.shp')
```

```python
gdf
```

```python
gdf = gdf[["mpio_cdpmp", "geometry"]]
```

```python
gdf = gdf.rename(columns={"mpio_cdpmp": "codigo_alt"})
```

```python
df = pd.read_csv('/content/base2p.csv')
```

```python
gdf = gdf.astype({"codigo_alt": "int"})
```

```python
m = df.merge(gdf, on="codigo_alt", how="inner")
```

```python
type(m)
```

```python
m = gpd.GeoDataFrame(m)
```

```python
type(m)
```

```python
m.to_parquet("pro2.parquet", index = False)
```
